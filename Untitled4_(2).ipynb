```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
from torchvision.datasets import MNIST
from torchvision import transforms
from collections import Counter
from sklearn.metrics import confusion_matrix, classification_report
import matplotlib.pyplot as plt
```

```python
# Load MNIST
transform = transforms.ToTensor()
train_dataset = MNIST(root="data", train=True, download=True, transform=transform)
test_dataset = MNIST(root="data", train=False, download=True, transform=transform)

X_train = train_dataset.data
y_train = train_dataset.targets
X_test = test_dataset.data
y_test = test_dataset.targets

# Make digit 0 minority
zero_idx = torch.where(y_train == 0)[0]
other_idx = torch.where(y_train != 0)[0]

keep_zeros = int(0.1 * len(zero_idx))  # keep only 10%
zero_idx = zero_idx[torch.randperm(len(zero_idx))[:keep_zeros]]

imbalanced_idx = torch.cat([zero_idx, other_idx])

X_train_imbalanced = X_train[imbalanced_idx]
y_train_imbalanced = y_train[imbalanced_idx]

print("Class distribution (imbalanced):", Counter(y_train_imbalanced.tolist()))
```

```python
zero_mask = y_train_imbalanced == 0
X_zero = X_train_imbalanced[zero_mask].float() / 255.0
X_zero_flat = X_zero.view(-1, 28*28)

zero_loader = DataLoader(TensorDataset(X_zero_flat), batch_size=64, shuffle=True)
```

```python
# Generator
class Generator(nn.Module):
    def __init__(self):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(100, 256),
            nn.ReLU(),
            nn.Linear(256, 512),
            nn.ReLU(),
            nn.Linear(512, 784),
            nn.Tanh()
        )
    def forward(self, z):
        return self.net(z)

# Discriminator
class Discriminator(nn.Module):
    def __init__(self):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(784, 512),
            nn.LeakyReLU(0.2),
            nn.Linear(512, 256),
            nn.LeakyReLU(0.2),
            nn.Linear(256, 1),
            nn.Sigmoid()
        )
    def forward(self, x):
        return self.net(x)

G = Generator()
D = Discriminator()

criterion = nn.BCELoss()
opt_G = optim.Adam(G.parameters(), lr=0.0002)
opt_D = optim.Adam(D.parameters(), lr=0.0002)
```

```python
epochs = 5          # keep small for Colab
noise_dim = 100
max_batches = 100   # limit batches per epoch

for epoch in range(epochs):
    for i, (real_imgs,) in enumerate(zero_loader):
        if i >= max_batches:
            break
        bs = real_imgs.size(0)
        real_imgs = real_imgs.view(bs, -1)

        real_labels = torch.ones(bs, 1)
        fake_labels = torch.zeros(bs, 1)

        # Train Discriminator
        z = torch.randn(bs, noise_dim)
        fake_imgs = G(z)
        loss_D = criterion(D(real_imgs), real_labels) + criterion(D(fake_imgs.detach()), fake_labels)
        opt_D.zero_grad()
        loss_D.backward()
        opt_D.step()

        # Train Generator
        z = torch.randn(bs, noise_dim)
        fake_imgs = G(z)
        loss_G = criterion(D(fake_imgs), real_labels)
        opt_G.zero_grad()
        loss_G.backward()
        opt_G.step()
    print(f"Epoch {epoch+1}/{epochs} done")
```

```python
# How many to generate
count = Counter(y_train_imbalanced.tolist())
majority_class = max(count, key=count.get)
minority_class = min(count, key=count.get)
need_samples = count[majority_class] - count[minority_class]

G.eval()
with torch.no_grad():
    z = torch.randn(need_samples, noise_dim)
    fake_imgs = G(z).view(-1, 1, 28, 28)
```

```python
# CNN-ready tensors
X_train_imbalanced_cnn = X_train_imbalanced.unsqueeze(1).float() / 255.0
X_test_cnn = X_test.unsqueeze(1).float() / 255.0

y_fake = torch.zeros(len(fake_imgs), dtype=torch.long)
X_train_vanilla_cnn = torch.cat([X_train_imbalanced_cnn, fake_imgs], dim=0)
y_train_vanilla = torch.cat([y_train_imbalanced, y_fake], dim=0)

print("Balanced dataset size:", X_train_vanilla_cnn.shape)
```

```python
class CNN(nn.Module):
    def __init__(self):
        super().__init__()
        self.conv = nn.Sequential(
            nn.Conv2d(1, 16, 3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2),
            nn.Conv2d(16, 32, 3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2)
        )
        self.fc = nn.Linear(32*7*7, 10)
    def forward(self, x):
        x = self.conv(x)
        x = x.view(x.size(0), -1)
        return self.fc(x)
```

```python
def train_classifier_fast(X_train, y_train, X_test, y_test, epochs=1):
    loader = DataLoader(TensorDataset(X_train, y_train), batch_size=32, shuffle=True)
    model = CNN()
    optimizer = optim.Adam(model.parameters(), lr=0.001)
    criterion = nn.CrossEntropyLoss()

    for _ in range(epochs):
        for xb, yb in loader:
            optimizer.zero_grad()
            loss = criterion(model(xb), yb)
            loss.backward()
            optimizer.step()

    model.eval()
    with torch.no_grad():
        preds = model(X_test).argmax(dim=1)
        print(confusion_matrix(y_test, preds))
        print(classification_report(y_test, preds))
```

```python
print("=== ORIGINAL IMBALANCED ===")
train_classifier_fast(X_train_imbalanced_cnn, y_train_imbalanced, X_test_cnn, y_test, epochs=1)
```

```python
# --------------------------
# Take small subset for fast Colab run
# --------------------------
subset_size = 2000   # can adjust, smaller = faster
X_train_fast = X_train_vanilla_cnn[:subset_size]
y_train_fast = y_train_vanilla[:subset_size]

X_test_fast = X_test_cnn[:1000]  # smaller test set
y_test_fast = y_test[:1000]
```

```python
print("=== VANILLA GAN BALANCED (FAST RUN) ===")
train_classifier_fast(X_train_fast, y_train_fast, X_test_fast, y_test_fast, epochs=1)
```

```python
# -------------------
# DCGAN Generator
# -------------------
class DCGAN_G(nn.Module):
    def __init__(self, noise_dim=100):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(noise_dim, 128),
            nn.ReLU(True),
            nn.Linear(128, 256),
            nn.ReLU(True),
            nn.Linear(256, 28*28),
            nn.Tanh()
        )
    def forward(self, z):
        return self.net(z)

# -------------------
# DCGAN Discriminator
# -------------------
class DCGAN_D(nn.Module):
    def __init__(self):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(28*28, 256),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Linear(256, 128),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Linear(128, 1),
            nn.Sigmoid()
        )
    def forward(self, x):
        return self.net(x)
```

```python
G_dcgan = DCGAN_G()
D_dcgan = DCGAN_D()

criterion = nn.BCELoss()
opt_G = optim.Adam(G_dcgan.parameters(), lr=0.0002)
opt_D = optim.Adam(D_dcgan.parameters(), lr=0.0002)

noise_dim = 100
epochs = 5
max_batches = 100  # limit for Colab speed

for epoch in range(epochs):
    for i, (real_imgs,) in enumerate(zero_loader):
        if i >= max_batches:
            break
        bs = real_imgs.size(0)
        real_imgs = real_imgs.view(bs, -1)

        real_labels = torch.ones(bs,1)
        fake_labels = torch.zeros(bs,1)

        # Train Discriminator
        z = torch.randn(bs, noise_dim)
        fake_imgs = G_dcgan(z)
        loss_D = criterion(D_dcgan(real_imgs), real_labels) + criterion(D_dcgan(fake_imgs.detach()), fake_labels)
        opt_D.zero_grad()
        loss_D.backward()
        opt_D.step()

        # Train Generator
        z = torch.randn(bs, noise_dim)
        fake_imgs = G_dcgan(z)
        loss_G = criterion(D_dcgan(fake_imgs), real_labels)
        opt_G.zero_grad()
        loss_G.backward()
        opt_G.step()
    print(f"DCGAN Epoch {epoch+1}/{epochs} done")
```

```python
# Decide how many images we need
count = Counter(y_train_imbalanced.tolist())
majority_class = max(count, key=count.get)
minority_class = min(count, key=count.get)
need_samples = count[majority_class] - count[minority_class]

G_dcgan.eval()
with torch.no_grad():
    z = torch.randn(need_samples, noise_dim)
    fake_imgs_dcgan = G_dcgan(z).view(-1, 1, 28, 28)

y_fake_dcgan = torch.zeros(len(fake_imgs_dcgan), dtype=torch.long)
```

```python
X_train_dcgan_cnn = torch.cat([X_train_imbalanced_cnn, fake_imgs_dcgan], dim=0)
y_train_dcgan = torch.cat([y_train_imbalanced, y_fake_dcgan], dim=0)

print("DCGAN balanced dataset size:", X_train_dcgan_cnn.shape)
```

```python
print("=== DCGAN BALANCED (FAST RUN) ===")
# Use small subset for speed
subset_size = 2000
X_train_dcgan_fast = X_train_dcgan_cnn[:subset_size]
y_train_dcgan_fast = y_train_dcgan[:subset_size]

X_test_fast = X_test_cnn[:1000]
y_test_fast = y_test[:1000]

train_classifier_fast(X_train_dcgan_fast, y_train_dcgan_fast, X_test_fast, y_test_fast, epochs=1)
```
